### Car Detections

```python
#car detections code
import cv2
from google.colab.patches import cv2_imshow
img_path='/content/drive/MyDrive/SM_Interaction/LMP/computer_vision/Computer_Vision_data/'
car_detector = cv2.CascadeClassifier(img_path+'/Cascades/cars.xml')
image = cv2.imread(img_path+'/Images/car.jpg')
img_converting_grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
detections = car_detector.detectMultiScale(img_converting_grey,scaleFactor=1.03,minNeighbors=5)
for(x,y,w,h) in detections:
   cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(image)
```
