**What id HOG ??**

The histogram of oriented gradients (HOG) is a feature descriptor used in computer vision and image processing for the purpose of object detection. The technique counts occurrences of gradient orientation in localized portions of an image. This method is similar to that of edge orientation histograms, scale-invariant feature transform descriptors, and shape contexts, but differs in that it is computed on a dense grid of uniformly spaced cells and uses overlapping local contrast normalization for improved accuracy.

```python
import dlib
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
```

#### Face Detection with HOG and Dlib

```python
img_path='/content/drive/MyDrive/SM_Interaction/LMP/computer_vision/Computer_Vision_data/'
image = cv2.imread(img_path+'/Images/people2.jpg')
face_detector_hog = dlib.get_frontal_face_detector()
detections = face_detector_hog(image,1)
#print(detections)
#print("Number of faces-----",len(detections))
for face in detections:
   #print(face.left())
   #print(face.top())
   #print(face.right())
   #print(face.bottom())
   l,t,r,b = face.left(),face.top(),face.right(),face.bottom()
   cv2.rectangle(image,(l,t),(r,b),(0,255,255),2)
cv2_imshow(image)
```

#### Haarcasecase Face Detection with CNN (Convolutional Neural Networks)

```python
img_path='/content/drive/MyDrive/SM_Interaction/LMP/computer_vision/Computer_Vision_data/'
image = cv2.imread(img_path+'/Images/people2.jpg')
cnn_detector = dlib.cnn_face_detection_model_v1(img_path+'Weights/mmod_human_face_detector.dat')
detections = cnn_detector(image,1)
for face in detections:
   l,t,r,b,c = face.rect.left(),face.rect.top(),face.rect.right(),face.rect.bottom(), face.confidence
   print(c)
   cv2.rectangle(image,(l,t),(r,b),(0,255,255),2)
cv2_imshow(image)
```

#### HARCASECDE & HOG & CNN

```python
#Harcasecade
img_path='/content/drive/MyDrive/SM_Interaction/LMP/computer_vision/Computer_Vision_data/'
image = cv2.imread(img_path+'/Images/people3.jpg')
image_grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
har_casecade_detector = cv2.CascadeClassifier(img_path+'/Cascades/haarcascade_frontalface_default.xml')
detections = har_casecade_detector.detectMultiScale(image_grey,scaleFactor=1.001,minNeighbors=5,minSize=(5,5))
for(x,y,w,h) in detections:
   cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(image)
#harcasecade is not a good options
```
